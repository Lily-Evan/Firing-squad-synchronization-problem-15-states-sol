Περιγραφή του προβλήματος Firing Squad Syncronization θα βρείτε στον παρακάτω σύνδεσμο:
Firing squad synchronization problem - Wikipedia
Η εφαρμογή αυτή πρέπει να αναπτυχθεί με τέτοιον τρόπο, ώστε η οπτικοποίηση των αρχικών
επιλογών και των αποτελεσμάτων να γίνεται μέσω ενός Graphic User Interface (GUI). Παραδοτέα
είναι ο κώδικας που θα αναπτυχθεί καθώς και μία αναφορά σχετικά λίγων σελίδων που να εξηγεί
την διαδικασία που ακολουθήθηκε και τις δυνατότητες του GUI που αναπτύχθηκε.

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

# Όλοι οι δυνατοί χαρακτήρες/καταστάσεις
states = {
    "·": 0,
    "G": 1,
    "L": 2,
    "R": 3,
    "A": 4,
    "B": 5,
    "C": 6,
    "a": 7,
    "b": 8,
    "c": 9,
    "bB": 10,
    "cC": 11,
    "RL": 12,
    "LR": 13,
    "F": 14
}
reverse_states = {v: k for k, v in states.items()}

# Συνάρτηση μετάβασης (ίδια όπως πριν)
def next(l, s, r):
    if s == "A":
        if r[0] == "L":
            return "G"
        return "B"
    if s == "B":
        return "C"
    if s == "C":
        if r[0] == "L":
            return "G"
        return "·"
    if s == "a":
        if l[-1] == "R":
            return "G"
        return "b"
    if s == "b":
        return "c"
    if s == "c":
        if l[-1] == "R":
            return "G"
        return "·"
    if s == "bB":
        return "cC"
    if s == "cC":
        if l[-1] == "R" or r[0] == "L":
            return "G"
        return "·"
    if s == "G":
        if (l + r) in ("$$", "$G", "G$", "GG"):
            return "F"
        if r == "·" and l == "·":
            return "bB"
        if r == "·":
            return "B"
        if l == "·":
            return "b"
    if s == "L":
        if l == "$" or l[-1] == "R":
            return "R"
        if l[-1] == "C":
            return "G"
        if r == "G":
            return "L"
        return "·"
    if s == "R":
        if r == "$" or r[0] == "L":
            return "L"
        if r[0] == "c":
            return "G"
        return "·"
    if s == "LR":
        if l[-1] == "C" or r[0] == "c":
            return "G"
        return "·"
    if s == "RL":
        return "LR"
    if s == "·":
        if l == "G" or l[-1] == "R":
            if r == "G" or r[0] == "L":
                return "RL"
            return "R"
        if r == "G" or r[0] == "L":
            if l == "G" or l[-1] == "R":
                return "RL"
            return "L"
        if l[-1] == "C":
            return "A"
        if r[0] == "c":
            return "a"
    return s

# Προετοιμασία animation
def simulate_firing_squad(n=13, max_steps=50):
    seq = ["·"] * n
    seq[0] = "G"
    states_over_time = [seq.copy()]

    for _ in range(max_steps):
        seq1 = []
        for j, s in enumerate(seq):
            l = seq[j - 1] if j - 1 >= 0 else "$"
            r = seq[j + 1] if j + 1 < len(seq) else "$"
            s1 = next(l, s, r)
            seq1.append(s1)
        states_over_time.append(seq1)
        seq = seq1
        if set(seq) == {"F"}:
            break
    return states_over_time

# Δημιουργία animation με matplotlib
def animate_firing_squad(seq_history):
    data = np.array([[states.get(cell, 0) for cell in row] for row in seq_history])

    fig, ax = plt.subplots(figsize=(12, 6))
    im = ax.imshow(data[:1], cmap="tab20", vmin=0, vmax=max(states.values()))

    def update(i):
        im.set_data(data[:i+1])
        return [im]

    ani = animation.FuncAnimation(fig, update, frames=len(data), interval=300, blit=True, repeat=False)
    plt.close(fig)
    return ani

# Run everything
history = simulate_firing_squad(n=21, max_steps=60)
ani = animate_firing_squad(history)

from IPython.display import HTML
HTML(ani.to_jshtml())
```
